```python
import numpy as np
import pandas as pd
import math
from pysat.formula import CNF
from pysat.solvers import Solver
import random
```

```python
data=pd.read_csv("sudoku.csv")
data
```

```python
num=np.random.randint(0,10000)
a=str(data['solutions'][num])
```

```python
sudoku1_sol=np.zeros([9,9],dtype=int)
```

```python
for i in range(9):
    for j in range(9):
        sudoku1_sol[i][j]=int(a[i*9+j])
        
sudoku1_sol
```

```python
def mod(n,m):
    if n%m==0:
        return m
    else:
        return n%m
```

```python
remaining_cells=[]
for i in range(1,9*9+1):
    remaining_cells.append(i)
    
flag=0

while(len(remaining_cells)>0):
    cell=random.choice(remaining_cells)
    cell_row=math.ceil(cell/9)-1
    cell_col=mod(cell,9)-1
    cell_val=sudoku1_sol[cell_row][cell_col]
    sudoku1_sol[cell_row][cell_col]=0

    for i in range(1,9+1):
        if i!=cell_val:
            cell_new_val=i
            sudoku1_sol[cell_row][cell_col]=i
        
            s=Solver()
        
            r0=[]
            for i in range(9):
                for j in range(9):
                    if sudoku1_sol[i][j]!=0:
                        for k in range(1,10):
                            if i*9*9+j*9+k!=i*9*9+j*9+sudoku1_sol[i][j]:
                                r0.append(-1*(i*9*9+j*9+k))
                            else:
                                r0.append(i*9*9+j*9+int(sudoku1_sol[i][j]))
            
                            
            li=[]
            cnf1=CNF()
            for i in range(1,9*9*9+1):
                li.append(i)
                if i%9==0:
                    cnf1.append(li)
                    s.add_clause(li)
                    li=[]
                
            cnf2=CNF()
            li1=[]
            for i in range(9):
                for j in range(9):
                    for k in range(1,10):
                        for l in range(k+1,10):
                            li1=[(-1*(i*9*9+j*9+k))]
                            li1.append(-1*(i*9*9+j*9+l))
                            cnf2.append(li1)
                            s.add_clause(li1)
                        
            r1=[]
            for i in range(1,(9*9)*(9*9)+1):
                a1=math.ceil(i/(9*9))
                b1=mod(i,9*9)
                row_a=math.ceil(a1/9)
                row_b=math.ceil(b1/9)
                col_a=mod(a1,9)
                col_b=mod(b1,9)
                block_row_a=math.ceil(row_a/3)
                block_col_a=math.ceil(col_a/3)
                block_row_b=math.ceil(row_b/3)
                block_col_b=math.ceil(col_b/3)
                if row_a==row_b:
                    r1.append(9*9*9+i)
                elif col_a==col_b:
                    r1.append(9*9*9+i)
                elif (block_row_a==block_row_b) and (block_col_a==block_col_b):
                    r1.append(9*9*9+i)
                 
            cnf3=CNF()
            li2=[]
            for i in r1:
                i=i-729
                a1=math.ceil(i/(9*9))
                b1=mod(i,9*9)
                row_a=math.ceil(a1/9)
                row_b=math.ceil(b1/9)
                col_a=mod(a1,9)
                col_b=mod(b1,9)
                block_row_a=math.ceil(row_a/3)
                block_col_a=math.ceil(col_a/3)
                block_row_b=math.ceil(row_b/3)
                block_col_b=math.ceil(col_b/3)
                for j in range(1,10):
                    if a1!=b1:
                        li2.append(-1*((row_a-1)*9*9+(col_a-1)*9+j))
                        li2.append(-1*((row_b-1)*9*9+(col_b-1)*9+j))
                        cnf3.append(li2)
                        s.add_clause(li2)
                        li2=[]
                        

                 
                
            final_assumption_list=[]
            for i in r0:
                final_assumption_list.append(int(i))
                
                
            if(s.solve(assumptions=final_assumption_list)):
                sudoku1_sol[cell_row][cell_col]=cell_val
                break
            sudoku1_sol[cell_row][cell_col]=0
    remaining_cells.remove(cell)

print(sudoku1_sol)
```
